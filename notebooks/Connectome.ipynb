```python
import boto3
import os

# Configure bucket and local destination
bucket_name = "open-neurodata"
prefix = "ara_2016/"
local_dir = "/home/adnan/Downloads/Connectome"

# Ensure the local directory exists
os.makedirs(local_dir, exist_ok=True)

# Create S3 client
s3 = boto3.client("s3")

# Paginator to handle large bucket sizes
paginator = s3.get_paginator("list_objects_v2")
pages = paginator.paginate(Bucket=bucket_name, Prefix=prefix)

# Download all files
print("Starting download...")
for page in pages:
    if "Contents" in page:
        for obj in page["Contents"]:
            key = obj["Key"]
            file_path = os.path.join(local_dir, key[len(prefix):])  # Strip the prefix from the path
            file_dir = os.path.dirname(file_path)
            os.makedirs(file_dir, exist_ok=True)  # Create directories if they don't exist

            # Download the file
            print(f"Downloading: {key} to {file_path}")
            s3.download_file(bucket_name, key, file_path)
    else:
        print("No files found in this page.")

print("Download complete.")
```
